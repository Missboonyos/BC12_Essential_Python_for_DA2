```python
## map / filter
## iterable
scores = [82, 55, 89, 72, 96, 49.5]
```

```python
for score in scores:
  if score >= 80:
    print("A")
  elif score >= 70:
    print("B")
  elif score >= 60:
    print("C")
  elif score >= 50:
    print("D")
  else:
    print("F")
```

```python
def grade(score):
  if score >= 90:
    return "A"
  elif score >= 80:
    return "B"
  elif score >=70:
    return "C"
  elif score >= 60:
    return "D"
  else:
    return "F"
```

```python
## map
list(map(grade, scores))
```

```python
## add five to scores
new_scores = []
for score in scores:
  new_scores.append(score + 5)
print(new_scores)
```

```python
## map, add five to scores
list(map(lambda x: x + 5, scores))
```

```python
students_score_over_80 = []

for score in scores:
  if score >= 80:
    students_score_over_80.append(score)

print(students_score_over_80)
```

```python
## filter
list(filter(lambda x: x >= 80, scores))
```

```python
## filter with tuple
scores = [ ("Jan", 88), ("Feb", 97), ("March", 63) ]

# filter score >= 80
list(filter(lambda x: x[1] >=80, scores))
```

```python
## module / library
## homework: class ATM
import random

class ATM:
  def __init__(self, name, balance):
    self.name = name
    self.balance = balance

  def __str__(self):
    return f"Account Name: {self.name}, Balance: EUR {self.balance}"

  def get_otp(self):
    otp = random.randint(1000,9999)
    return f"Your requested OTP is {otp}"
```

```python
my_account = ATM("Jaykin", 10000)
```

```python
print(my_account)
```

```python
## ATM method
my_account.get_otp()
```
